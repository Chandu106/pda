```python
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.mplot3d import Axes3D

# Create a figure and a 3D axis
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Generate data for the x and y variables
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)

# Define a function to calculate the z values
def f(x, y):
    return np.sin(np.sqrt(x**2 + y**2))

# Calculate the z values using the function
Z = f(X, Y)

# Create the 3D surface plot
surf = ax.plot_surface(X, Y, Z, cmap='viridis')

# Customize the plot
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title('3D Surface Plot')
fig.colorbar(surf, ax=ax, shrink=0.5, aspect=5)

# Show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np

# Sample data for the 3D bar chart
x = [1, 2, 3, 4, 5]
y = [1, 2, 3, 4, 5]
z = [1, 2, 3, 4, 5]

# Create a 3D bar chart
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.bar(x, y, z, zdir='y', color='b', alpha=0.8)

# Customize the plot
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_zlabel('Z-axis')
ax.set_title('3D Bar Chart')

# Show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0,5,100)
y1 = np.sin(x)
y2 = np.cos(x)
y3 = np.tan(x)
y4 = np.log(x)

plt.figure(figsize=(10,7))
plt.subplot(2,2,1)
plt.plot(x,y1,c='r')
plt.title("Sine")

plt.subplot(2,2,2)
plt.plot(x,y2,c='g')
plt.title("Cosine")

plt.subplot(2,2,3)
plt.plot(x,y3,c='b')
plt.title("Tangent")

plt.subplot(2,2,4)
plt.plot(x,y4,'b')
plt.title('loagarthmic')

plt.tight_layout()
plt.show()
```

```python
import numpy as np
x1 = np.array([0,1,2,3,4,5])
y1 = np.array([0,100,200,300,400,500])

x2 = np.array([0,1,2,3,4,5])
y2 = np.array([50,20,40,20,60,70])

x3 = np.array([0,1,2,3,4,5])
y3 = np.array([10,20,30,40,50,60])

x4 = np.array([0,1,2,3,4,5])
y4 = np.array([200,350,250,550,450,150])

plt.figure(figsize=(10,7))
plt.subplot(2,2,1)
plt.plot(x1,y1, marker='o',linestyle='--',color='r')
plt.xlabel("x1 label")
plt.ylabel("y1 label")
plt.title("x1 vs y1")

plt.subplot(2,2,2)
plt.plot(x2,y2,marker='o',linestyle='--',color='r')
plt.xlabel("x2 label")
plt.ylabel("y2 label")
plt.title("x2 vs y2")

plt.subplot(2,2,3)
plt.plot(x3,y3,marker='o',linestyle='--',color='r')
plt.xlabel("x3 label")
plt.ylabel("y3 label")
plt.title("x3 vs y3")

plt.subplot(2,2,4)
plt.plot(x4,y4,marker='o',linestyle='--',color='r')
plt.xlabel("x4 label")
plt.ylabel("y4 label")
plt.title("x4 vs y4")

plt.legend()
plt.tight_layout()
plt.show()
```
