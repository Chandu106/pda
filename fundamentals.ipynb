```python
company_sales_data = pd.read_csv("https://raw.githubusercontent.com/Chandu106/pda/main/expt9/company_sales_data%20-%20company_sales_data.csv")
company_sales_data.head()
```

```python
month_number = company_sales_data["month_number"]
month_number
```

```python
total_profits = company_sales_data["total_profit"]
total_profits
```

```python
plt.plot(month_number, total_profits)
plt.xlabel("Month Number")
plt.ylabel("total profits")
plt.title("month vs profits")
plt.grid(True)
plt.legend()
plt.show()
```

```python
plt.plot(month_number, total_profits, marker='o', alpha=0.8, linestyle='--', color='red')
plt.xlabel("month")
plt.ylabel("profits")
plt.title("months vs profits")
plt.legend(loc="lower right")
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt

# Sample data for two lines
x = [1, 2, 3, 4, 5]
y1 = [2, 4, 6, 8, 10]
y2 = [1, 3, 5, 7, 9]

# Create a figure and axis
fig, ax = plt.subplots()

# Plot the first line
ax.plot(x, y1, label='Line 1', color='blue', linestyle='-', marker='o')

# Plot the second line
ax.plot(x, y2, label='Line 2', color='red', linestyle='--', marker='s')

# Add labels and legend
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_title('Multiple Lines in a Single Plot')
ax.legend()

# Display the plot
plt.show()
```
