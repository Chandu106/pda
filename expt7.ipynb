# **Reading the Data**

```python
import pandas as pd
automobile_data = pd.read_csv("https://raw.githubusercontent.com/Chandu106/pda/main/expt7/Automobile_data%20-%20Automobile_data.csv")
```

# **2.Find the most expensive car company name**


```python
price = automobile_data["price"]
company = automobile_data["company"]
index_highest_price = price.idxmax()
print("Highest car price: ",price[index_highest_price])
print("Index of the car price with highest cost: ",index_highest_price)
print("The most expensive car company: ", company[index_highest_price])
```

# **3.Print All Toyota Cars details**

```python
toyota_car_company = automobile_data[automobile_data.company == "toyota"]
toyota_car_company
```

# **4.Count total cars per company**

```python
car_count = company.value_counts()
car_count
```

# **5.Find each company’s Highest price car**

```python
highest_price = automobile_data.groupby(company)['price'].max()
highest_price
```

# **6.Find the average mileage of each car making company**

```python
average_mileage = automobile_data.groupby(company)['average-mileage'].mean()
average_mileage
```

# **7.Print the unique makers**


```python
unique_car_makers = company.unique()
unique_car_makers
```
