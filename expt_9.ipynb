# **Reading the Data**

```python
import matplotlib.pyplot as plt
import pandas as pd
company_sales_data = pd.read_csv("https://raw.githubusercontent.com/Chandu106/pda/main/expt9/company_sales_data%20-%20company_sales_data.csv")
```

```python
company_sales_data.head()
```

# **1.Read Total profit of all months and show it using a line plot**

```python
total_profit = company_sales_data["total_profit"]
plt.plot(total_profit)
plt.xlabel("Months")
plt.ylabel("Profit")
plt.title("TOTAL PROFIT")
```

# **2.Get total profit of all months and show line plot with the following Style properties Generated line plot must include following Style properties Line Style doted and Line-color should be red, Show legend at the lower right location. X label name = Month Number,Y label name = Sold units number,Add a circle marker.,Line marker color as read, Line width should be 3**

```python
total_profit = company_sales_data["total_profit"]
months = company_sales_data["month_number"]
plt.plot(months, total_profit, linestyle='dotted', color='red', marker='o', markerfacecolor='red', markersize=9, linewidth=3)
plt.xlabel("Month Number")
plt.ylabel("Sold units number")
plt.title("Months vs sold units")
plt.legend(loc="lower right")
```

# **4.Read face cream and facewash product sales data and show it using the bar chart**

# **5.Calculate total sale data for last year for each product and show it using a Pie chart**


```python
total_sales = {
    'Face Cream': facecream.sum(),
    'Face Wash': facewash.sum(),
    'Toothpaste': toothpaste.sum(),
    'Bathing Soap': bathingsoap.sum(),
    'Shampoo': shampoo.sum(),
    'Moisturizer': moisturizer.sum()
}
products = list(total_sales.keys())
sales = list(total_sales.values())
plt.figure(figsize=(8, 8))
plt.pie(sales, labels=products, autopct='%1.1f%%', startangle=140)
plt.title('Total Sales for 6 Products')
```

# **6.Read Bathing soap facewash of all months and display it using the Subplot**

```python
# Create subplots
fig, axes = plt.subplots(1, 2, figsize=(12, 5))

# Plot Bathing soap sales
axes[0].bar(months, bathingsoap, color='skyblue')
axes[0].set_title('Bathing Soap Sales')
axes[0].set_xlabel('Month')
axes[0].set_ylabel('Sales')

# Plot Facewash sales
axes[1].bar(months, facewash, color='lightcoral')
axes[1].set_title('Facewash Sales')
axes[1].set_xlabel('Month')
axes[1].set_ylabel('Sales')

# Adjust spacing between subplots
plt.tight_layout()

# Show the subplots
plt.show()
```
